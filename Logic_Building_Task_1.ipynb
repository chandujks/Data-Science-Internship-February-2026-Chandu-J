```python
# 1. user login check

correct_username = "admin"
correct_password = "1234"

username = input("Enter username: ")
password = input("Enter password:")

if username == correct_username and password == correct_password:
    print("login succesful")
else:
        print("Invalid Credentials")
```

```python
# 2.pass/fail analyzer

marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Total Students:", len(marks))
print("Number of Pass Students:", pass_count)
print("Number of Fail Students:", fail_count)
```

```python
# 3. simple data cleaner

names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []


for name in names:
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)


print("Cleaned Names:", cleaned_names)
```

```python
# 4.message length analyzer

messages = ["Hi", "Welcome to the platform", "OK"]


for message in messages:
    length = len(message)
    print("Message:", message)
    print("Length:", length)

    if length > 10:
        print("Flag: Message is longer than 10 characters")

    print()
```

```python
# 5. error message detector

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

# Count ERROR entries
for log in logs:
    if log == "ERROR":
        error_count += 1

print("Number of ERROR entries:", error_count)
```
