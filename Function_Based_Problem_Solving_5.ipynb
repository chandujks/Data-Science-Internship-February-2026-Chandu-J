```python
# 1. Social media post engagement analyzer

# Function to analyze post engagement
def analyze_engagement(likes_list):

    total_likes = 0

    # Calculate total likes using loop
    for like in likes_list:
        total_likes += like

    # Check post status
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    # Display output
    print("Total Likes:", total_likes)
    print("Post Status:", status)

# Example likes data
likes = [200, 300, 250, 300]

# Calling function
analyze_engagement(likes)
```

```python
# 2 . Healthcare medicine stock alert system

# Function to monitor medicine stock
def check_medicine_stock(stock):

    # Check stock level using condition
    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"

    # Display output
    print("Medicine Stock:", stock)
    print("Status:", status)

# Example stock value
medicine_stock = 6

# Calling function
check_medicine_stock(medicine_stock)
```

```python
# 3. Rainfall adequacy checker

# Function to check rainfall adequacy
def check_rainfall(rainfall_data, required_level):

    total_rainfall = 0

    # Calculate total rainfall using loop
    for rain in rainfall_data:
        total_rainfall += rain

    # Calculate average rainfall
    average = total_rainfall / len(rainfall_data)

    # Check rainfall adequacy
    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    # Display output
    print("Average Rainfall:", average)
    print("Rainfall Status:", status)

# Example rainfall data (mm)
seasonal_rainfall = [60, 75, 80, 73]

required_rainfall = 70

# Calling function
check_rainfall(seasonal_rainfall, required_rainfall)
```

```python
# 4 . Duplicate account detection in social media

# Function to detect duplicate accounts
def check_duplicates(usernames):

    # Convert list to set
    unique_users = set(usernames)

    # Compare lengths
    if len(unique_users) < len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

# Example username list
users = ["john", "alice", "john", "mike"]

# Calling function
check_duplicates(users)
```

```python
# 5. Appointment eligiblity checker in healthcare system

# Function to check appointment eligibility
def check_appointment_eligibility(age):

    # Check age condition
    if age >= 18:
        status = "Eligible"
    else:
        status = "Not Eligible"

    # Display output
    print("Patient Age:", age)
    print("Eligibility Status:", status)

# Example patient age
patient_age = 21

# Calling function
check_appointment_eligibility(patient_age)
```

```python
# 6. Premium crop price filter in agriculture

# Function to filter premium crop prices
def filter_premium_crops(crop_prices):

    premium_crops = []

    # Filter prices using loop
    for price in crop_prices:
        if price > 2000:
            premium_crops.append(price)

    # Display premium crops
    print("Premium Crops:", premium_crops)

# Example crop prices
prices = [1500, 2500, 1800, 3200]

# Calling function
filter_premium_crops(prices)
```

```python
# 7 . System Monitoring-application health checker

# Function to evaluate application health
def check_system_health(error_count):

    # Check error conditions
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    # Display output
    print("Error Count:", error_count)
    print("System Status:", status)

# Example error count
errors = 5

# Calling function
check_system_health(errors)
```

```python
# 8 . Daily transaction limit checker in Bbanking

# Function to validate daily transaction
def check_transaction_limit(amount):

    daily_limit = 50000

    # Check transaction amount
    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"

    # Display output
    print("Transaction Amount:", amount)
    print("Transaction Status:", status)

# Example transaction amount
transaction_amount = 40000

# Calling function
check_transaction_limit(transaction_amount)
```

```python
# 9. Student attendence eligiblity in E learning

# Function to check exam eligibility
def check_exam_eligibility(attendance_list):

    present_count = 0

    # Count present days using loop
    for status in attendance_list:
        if status == "P":
            present_count += 1

    # Calculate attendance percentage
    total_days = len(attendance_list)
    percentage = (present_count / total_days) * 100

    # Check eligibility
    if percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"

    # Display output
    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", eligibility)

# Example attendance list
attendance = ["P", "P", "A", "P", "P","P"]

# Calling function
check_exam_eligibility(attendance)
```

```python
# 10. Smart electricty bill analyzer

# Function to calculate electricity bill
def calculate_electricity_bill(units):

    bill = 0

    # Billing Slabs
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage Classification
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    return bill, usage

# Example units consumed
units_consumed = 250

# Calling function
total_bill, usage_status = calculate_electricity_bill(units_consumed)

print("Total Bill: ₹", total_bill)
print("Usage Level:", usage_status)
```
