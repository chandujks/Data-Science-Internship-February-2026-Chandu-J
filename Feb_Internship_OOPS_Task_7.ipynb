```python
# 1 . Smart light controller

# Smart Light Class
class SmartLight:

    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    # for turn  ON light
    def turn_on(self):
        self.status = "ON"
        print(f"{self.name} is {self.status}")

    # for turn  OFF light
    def turn_off(self):
        self.status = "OFF"
        print(f"{self.name} is {self.status}")

    # for to display the status
    def show_status(self):
        print(f"{self.name} is currently {self.status}")



light_name = "Living area Light"
action = "ON"

# Creating object
living_area_light = SmartLight(light_name)

# Performing action
if action == "ON":
   living_area_light.turn_on()
else:
    living_area_light.turn_off()
```

```python
# 2 .Employee ID card system

# Employee ID Card Class
class Employee:

    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    # for display the ID card details
    def display_id_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


# Sample Input
employee_name = "Chandu"
employee_id = "EMP105"
department = "IT"

# Createing object
emp = Employee(employee_name, employee_id, department)

# Display ID Card
emp.display_id_card()
```

```python
# 3 . Mobile Contact Record

# Contact class
class Mobile_Contact:

    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    # for to display contact
    def display_mobile_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


#  Input
contact_name = "Anita"
phone_number = "9876543210"

# Createing  contact object
my_mobile_contact = Mobile_Contact(contact_name, phone_number)

# Displaying contact details
my_mobile_contact.display_mobile_contact()
```

```python
#  4. Product Price Tag Generator

# Class to represent a product
class Product:

    def __init__(self, name, price):
        self.name = name
        self.price = price

    # for to print price tag
    def print_price_tag(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))


# Input
product_name = "Sony Headphones"
product_price = 24999

# product object
item = Product(product_name, product_price)

# Printing  price tag
item.print_price_tag()
```

```python
# 5.  Movie Rating Display System

# class for a movie
class Movie:

    def __init__(self, title, rating):
        self.title = title
        self.rating = rating

    # for to display movie details
    def show_details(self):
        print("Movie:", self.title)
        print("Rating:", self.rating, "/ 5")


# Input
movie_name = "Inception"
movie_rating = 4.8

#  movie object
my_movies = Movie(movie_name, movie_rating)

# Display details
my_movies.show_details()
```

```python
# 6 . Delivery Address Manager

# Class delivery
class Delivery:

    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    # for to print delivery details
    def print_details(self):
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)


#  Input
name = "Suman"
location = "Hyderabad"

# Createing order object
order = Delivery(name, location)

# Displaying  delivery information
order.print_details()
```
