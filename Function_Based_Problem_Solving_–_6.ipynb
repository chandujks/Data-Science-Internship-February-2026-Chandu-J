```python
# 1 . Smart parking lot management system

# Function to manage parking system
def manage_parking(capacity, logs):

    current_vehicles = 0
    peak_usage = 0

    # Process each log entry
    for log in logs:

        if log == "IN":
            current_vehicles += 1
        elif log == "OUT":
            current_vehicles -= 1

        # Track peak usage
        if current_vehicles > peak_usage:
            peak_usage = current_vehicles

    # Check parking status
    if current_vehicles > capacity:
        status = "Parking Full! Alert!"
    else:
        status = "Available"

    # Display output
    print("Currently Parked Vehicles:", current_vehicles)
    print("Peak Usage:", peak_usage)
    print("Parking Status:", status)

# Sample Input
parking_capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

# Calling function
manage_parking(parking_capacity, vehicle_logs)
```

```python
# 2. Online food delivery time estimator

# Function to estimate delivery time
def estimate_delivery_time(distance, traffic, weather):

    # Base delivery time
    eta = distance * 5   # 5 minutes per km

    # Traffic delay
    if traffic == "High":
        eta += 15

    # Weather delay
    if weather == "Rainy":
        eta += 10

    print("Estimated Delivery Time:", eta, "minutes")

# Sample Input
distance = 8
traffic = "High"
weather = "Rainy"

# Function Call
estimate_delivery_time(distance, traffic, weather)
```

```python
# 3 . Movie theatre seat dccupancy analyzer

# Function to analyze seat occupancy
def analyze_occupancy(total_seats, booked_seats):

    # Count booked seats
    booked_count = len(booked_seats)

    # Calculate occupancy percentage
    occupancy = (booked_count / total_seats) * 100

    # Determine show status
    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Seats Available"

    # Suggest additional show
    if occupancy >= 75:
        suggestion = "Consider Opening Additional Show"
    else:
        suggestion = "No Additional Show Needed"

    # Display output
    print("Occupancy:", int(occupancy), "%")
    print("Show Status:", status)
    print("Suggestion:", suggestion)

# Sample Input
total_seats = 200
booked_seats = [1] * 150   # 150 seats booked

# Calling function
analyze_occupancy(total_seats, booked_seats)
```

```python
#  4 . Cloud server load classification system

# Function to classify server load
def classify_server_load(cpu_readings):

    total = 0

    # Calculate total using loop
    for reading in cpu_readings:
        total += reading

    # Calculate average CPU load
    average = total / len(cpu_readings)

    # Classify server status
    if average < 50:
        status = "Normal"
    elif average <= 80:
        status = "Warning"
    else:
        status = "Critical"

    # Display output
    print("Average CPU Load:", int(average), "%")
    print("Server Status:", status)

# Sample Input
cpu_data = [45, 60, 70, 85, 90]

# Calling function
classify_server_load(cpu_data)
```

```python
# 5 . Smart lcassroom resource usage monitor

# Function to monitor classroom resource usage
def monitor_resources(resource_usage):

    overused = []

    # Check each resource usage
    for resource, hours in resource_usage.items():
        if hours > 8:
            overused.append(resource)

    # Check energy alert
    if len(overused) > 0:
        alert = "Yes"
    else:
        alert = "No"

    # Display output
    print("Overused Resources:", ", ".join(overused))
    print("Energy Alert:", alert)

# Sample Input
usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

# Calling function
monitor_resources(usage)
```

```python
# 6 . Online event registration capacity controller

# Function to manage event registrations
def manage_registrations(capacity, registrations):

    confirmed = 0
    waitlisted = 0

    # Process each registration
    for i in range(registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    # Check registration status
    if confirmed >= capacity:
        status = "Closed"
    else:
        status = "Open"

    # Display output
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)

# Sample Input
event_capacity = 100
total_registrations = 105

# Calling function
manage_registrations(event_capacity, total_registrations)
```
