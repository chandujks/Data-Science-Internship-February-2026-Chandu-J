```python
# 1. Function to validate  mobile recharge
def validate_mobile_recharge():

    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: ₹"))

        # check minimum recharge rule
        if amount < 50:
            print("Recharge amount must be at least ₹50. Try again.\n")
            continue

        # Check if amount is a valid plan
        if amount in valid_plans:
            print("Recharge Successful!")
            break
        else:
            print("Invalid Plan! Available Plans: 199, 299, 399, 599")
            print("Please try again.\n")

# calling the function
validate_mobile_recharge()
```

```python
# 2.Inventory record alert system

# Function to check inventory stock
def check_inventory(inventory):

    # loop through dictionary
    for product, stock in inventory.items():

        # Check stock level
        if stock < 15:
            print(f"{product}: Reorder Alert (Stock: {stock})")
        else:
            print(f"{product}: Stock OK (Stock: {stock})")

# inventory data
products = {
    "Mobile": 10,
    "Laptop": 25,
    "Headphones": 12,
    "Keyboard": 18,
    "Moniter": 23,
    "Mouse": 15.
}

# call the function
check_inventory(products)
```

```python
# 3. Student result processing system

# Function to process student result
def process_result(marks):

    total = 0

    # calculate total using loop
    for mark in marks:
        total += mark

    # calculate average
    average = total / len(marks)

    # determine pass/fail
    if average >= 50:
        result = "Pass"
    else:
        result = "Fail"

    return result

#  student marks
student_marks = [60, 55, 40, 70]

# calling function
final_result = process_result(student_marks)

print("Result:", final_result)
```

```python
# 4. Cab fare estimator with retry option

# Function to calculate cab fare
def calculate_fare(distance, peak):

    base_fare = 50
    per_km_rate = 12

    # calculate normal fare
    fare = base_fare + (distance * per_km_rate)

    # add 25% extra during peak hour
    if peak.lower() == "yes":
        fare += fare * 0.25

    return fare

# retry option using while loop
while True:

    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ")

    total_fare = calculate_fare(distance, peak)

    print(f"Total Fare: ₹{total_fare}")

    retry = input("Do you want to calculate again? (yes/no): ")

    if retry.lower() != "yes":
        print("Thank you for using Cab Fare Estimator!")
        break
```

```python
# 5 . Employee attendence eligiblity checker

# Function to check attendance eligibility
def check_eligibility(attendance):

    present_days = 0

    # count present days using loop
    for status in attendance:
        if status == "P":
            present_days += 1

    # calculate percentage
    total_days = len(attendance)
    percentage = (present_days / total_days) * 100

    # check eligibility
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

# given attendance list
attendance_list = ["P", "P", "A", "P", "P"]

# calling function
result = check_eligibility(attendance_list)

print("Eligibility Status:", result)
```

```python
# 6. Password strength checker

# Function to check password strength
def check_password_strength(password):

    has_digit = False
    has_special = False
    special_chars = "@#$"

    # check minimum length
    if len(password) < 8:
        return "Weak Password"

    # loop through each character
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    # Check all conditions
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password prone to attacks"

# taking  input
pwd = input("Enter your password: ")

# calling function
result = check_password_strength(pwd)

print("Password Strength:", result)
```
