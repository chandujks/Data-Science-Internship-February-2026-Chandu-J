```python
# 1. Employee performance bonous eligibility
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
highest_score = max(employees.values())

top_performers = []

for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

print("Top Performers Eligible for Bonus:",
      ", ".join(top_performers),
      f"(Score: {highest_score})")
```

```python
# 2.Search Query keyword analysis
import string

text = "Buy mobile phone buy phone online"

# Step 1: Convert to lowercase
text = text.lower()

# Step 2: Remove punctuation
text = text.translate(str.maketrans('', '', string.punctuation))

# Step 3: Split into words
words = text.split()

# Step 4: Count frequency
frequency = {}

for word in words:
    frequency[word] = frequency.get(word, 0) + 1

# Step 5: Filter words appearing more than once
result = {}

for word, count in frequency.items():
    if count > 1:
        result[word] = count

print(result)
```

```python
# 3. Sensor data validation
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

# Loop with index
for hour, value in enumerate(sensor_readings):
    if value % 2 == 0:   # Check if even (valid)
        valid_readings.append((hour, value))

# Print result
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
# 4. Email domain usage analysis

# Given email list
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

# Step 1: Extract domains and count
for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

# Step 2: Calculate percentage
total_users = len(emails)

print("Email Domain Usage:")
for domain, count in domain_count.items():
    percentage = (count / total_users) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
# 5.Sales spike detection spike

sales = [1200, 1500, 900, 2200, 1400, 3000]

average_sales = sum(sales) / len(sales)

threshold = average_sales * 1.30

for day, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {day}: {value}")
```

```python
# 6. Duplicate user id detection

# Given list of user IDs
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Dictionary to store frequency
user_count = {}

# Step 1: Count occurrences
for user in user_ids:
    user_count[user] = user_count.get(user, 0) + 1

# Step 2: Display duplicates only
for user, count in user_count.items():
    if count > 1:
        print(f"{user} → {count} times")
```
